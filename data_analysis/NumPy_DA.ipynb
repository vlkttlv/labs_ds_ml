#### 1.4 Создайте функцию, принимающую на вход размеры двумерного массива и возвращающую  массив заданного размера с единицами на границах и нулями внутри

```python

def change_array(rows: int, cols: int):
    array = np.zeros((rows, cols))
    array[0], array[rows-1] = 1, 1   # меняем строки на единичные
    array[:,0], array[:,rows] = 1, 1 # меняем столбцы на единичные
    return array
print(change_array(4,5))
```

 Создайте одномерный массив numpy со случайными значениями и вычислите его стандартное отклонение по следующей формуле:

Для этого создайте функцию, которая:
 1. Рассчитывает среднее значение элементов массива
 2. Рассчитывает сумму квадратов разницы элементов массива и среднего значения
 3. Рассчитывает отношение полученной суммы к числу элементов массива
 4. Рассчитывает квадратный корень полученного значения

 Сравните вашу функцию и функцию библиотеки numpy std()

```python
array = np.random.random(5) 
def sigma(array):
    array_mean = np.mean(array)                         # среднее значение элементов массива
    sum_of_squares = np.sum((array - array_mean) ** 2)  # сумма квадратов разницы элементов массива и среднего значения
    variance = sum_of_squares / np.size(array)          # отношение полученной суммы к числу элементов массива
    return np.sqrt(variance)                            # квадратный корень полученного значения
    
    
my_res = sigma(array)
res_std = np.std(array)
print(f'my res: {my_res}')
print(f'numpy res: {res_std}')
```
