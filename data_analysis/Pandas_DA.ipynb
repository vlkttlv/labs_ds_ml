18. Выведите, сколько раз встречается значение каждой категории столбца 'Make' с помощью метода `value_counts()`

    Назовите полученный датафрейм 'counts'

```python
# counts = df['Make'].value_counts()
# counts = pd.DataFrame({'Make':counts.index, 'Count':counts.values})
# counts
counts = df['Make'].value_counts()
counts
```

Часть наших признаков имеют слишком много записей. Редкие из них можно исключить без потери точности. 

Сделаем эту операцию объявив такие записи одним типом - rare.



19. Посчитайте медиану встречаемости категорий counts (метод `.median()`)

```python
counts_median = counts.median()
counts_median
```

20. Замените в столбце 'Make' датафрейма df категории, которые встречаются реже медианного значения раз, на категорию 'Rare'

```python
for idx, row in df.iterrows():
    if counts[df['Make'].iloc[idx]] < counts_median:
        df['Make'].iloc[idx] = 'Rare'
df
```

## Бонусные задачи

__22*. Какая цена у самой дешевой машины от Toyota с автоматическим типом трансмиссии__

```python
price = df[(df['Make'] == 'Toyota') & (df['Transmission'] == 'Automatic')].sort_values('Price(euro)').iloc[0]['Price(euro)']
print(f"самая дешевая тойота стоит: {price} евро")
```

__23*. Какую пропорцию составляют машины с автоматической трансмиссией среди всех машин производителя Toyota__

```python
# df[(df['Make'] == 'Toyota') & (df['Transmission'] == 'Automatic')].count()
# находим сколько тайот с автоматической миссией
one = df[(df['Make'] == 'Toyota') & (df['Transmission'] == 'Automatic')].count().iloc[0] 
two = df[(df['Make'] == 'Toyota')].count().iloc[0] # находим сколько всего тайот
print(f"proportion: {one/two}")
```

__24*. Найдите среднее и стандартное отклонение для объёма двигателя машин Volkswagen с 2000 по 2010 года__

```python
v_df = df[(df['Make'] == 'Volkswagen') & (df['Year'] < 2010) & (df['Year'] > 2000)]
print(f"mean: {v_df['Engine_capacity(cm3)'].mean()}")
print(f"std: {v_df['Engine_capacity(cm3)'].std()}")
```

```python
merc_df = df[(df['Make'] == 'Mercedes') &  (df['Transmission'] == 'Manual')].drop_duplicates('Model')
print("уникальные модели мерса: ")
for _, row in merc_df.iterrows():
    print(row['Model'])
print(f"всего уникальностей: {merc_df['Model'].count()}")

```

__26*. Найдите самый многочисленный и самый малочисленный тип топлива у машин__

__27*. Найдите абсолютную разницу между количеством уникальных моделей автомобилей Renault и Toyota__

```python
count_reno = df[df['Make'] == 'Renault']['Model'].nunique()
print(f"количество уникальных моделей рено: {count_reno}")
count_toyota = df[df['Make'] == 'Toyota']['Model'].nunique()
print(f"количество уникальных моделей тойоты: {count_toyota}")
print(f"абсолютная разница: {abs(count_reno-count_toyota)}")
```
