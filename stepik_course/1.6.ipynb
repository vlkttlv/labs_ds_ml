```python
import pandas as pd
import numpy as np
```

```python
students_perfomance = pd.read_csv('datasets/StudentsPerformance.csv')
students_perfomance = students_perfomance.rename(columns={'math score':'math_score',
     'writing score':'writing_score',
     'reading score':'reading_score'})
```

```python
students_perfomance.head()
```

```python
students_perfomance.groupby(['gender', 'race/ethnicity'], as_index=False).aggregate({'math_score': 'mean'}).rename(columns={'math score': 'mean_math_score'})
# students_perfomance.groupby('gender').aggregate({'math score': 'mean'})
```

```python
scores = students_perfomance.groupby(['gender', 'race/ethnicity']).aggregate(
    {'math_score': 'mean', 'writing_score':'mean'}).rename(
        columns={'math_score': 'mean_math_score',
                 'writing_score': 'mean_writing_score'})
```

```python
scores.loc[[('female', 'group A'), ('female', 'group B')]]
```

```python
students_perfomance.groupby(['gender', 'race/ethnicity']).math_score.nunique()
```

```python
# топ математиков девушек и парней
students_perfomance.sort_values(['gender', 'math_score'],
                                ascending=False).groupby(['gender']).head()
```

```python
total_score = students_perfomance.math_score + students_perfomance.reading_score + students_perfomance.writing_score
students_perfomance['total_score'] = total_score
students_perfomance.head()
```

```python
students_perfomance = students_perfomance.assign(total_score_log = np.log(students_perfomance.total_score))
```

```python
students_perfomance.head()
```

```python
students_perfomance.drop(['total_score'], axis=1)
```
