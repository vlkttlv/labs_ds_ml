```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree # модуль для обучения решающих деревьев и использовать их для предсказания, обучения моделей 
```

```python
data = pd.DataFrame({'X_1': [1, 1, 1, 0, 0, 0, 0, 1],
                     'X_2': [0, 0, 0, 1, 0, 0, 0, 1],
                     'Y': [1, 1, 1, 1, 0, 0, 0, 0]})
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy') # дерево решений
```

```python
X = data[['X_1', 'X_2']] # фичи
y = data.Y # целевая переменная
```

```python
clf.fit(X, y) # обучаем
```

```python
tree.plot_tree(clf, filled=True, class_names=['Negative', 'Positive'], feature_names=list(X))
```
